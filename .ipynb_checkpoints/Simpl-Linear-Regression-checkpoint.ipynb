```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('height-weight.csv')
```

```python
df.head()
```

```python
#Scatter plot
plt.scatter(df['Weight'], df['Height'])
plt.title("Height - Weight Scatter Plot")
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
#Correlation

df.corr()
```

```python
import seaborn as sns
sns.pairplot(df)
plt.show()
```

```python
# Independent and dependent features
x = df[['Weight']] ###Independent feature should be a Dataframe or two dimensional array
y = df['Height'] ##Dependent feature should be a Series or one diemensional array
```

```python
#Train test splitabs
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42) 
```

```python
x_train.shape
```

```python
##Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
```

```python
#To prevent data leakage we only use transform ins
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
#Apply linear regression

from sklearn.linear_model import LinearRegression

```

```python
regression = LinearRegression()
```

```python
regression.fit(x_train, y_train)
```

```python
print(f"Coefficient or slope is : {regression.coef_} ")
print(f"Intercept is : {regression.intercept_} ")
```

```python
## Plot best fit line using training data
```

```python
plt.scatter(x_train, y_train)
plt.show()
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, regression.predict(x_train))
plt.show()
```

```python
#Prediction for test data 
```

```python
y_pred = regression.predict(x_test)    
```

```python
#Performance metricsabs
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("MSE", mse)
print("MAE", mae)
print("RMSE", rmse)
```

#R-SQAURE

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test, y_pred)
print(score)
```
